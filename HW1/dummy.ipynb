```python
import importlib
from table import Table
from dummy_model import DummyModel
import evaluations
import datetime
```

***Making Predictions from the Dataset***

```python
stroke_data_table = Table()
stroke_data_table.load_file("full_data.csv")
dummy = DummyModel()
attr_headers = ["gender", "age", "hypertension", "heart_disease", "ever_married", "work_type", "Residence_type", "avg_glucose_level", "bmi", "smoking_status"]
attr_data = stroke_data_table.make_sub_table(stroke_data_table, attr_headers)
class_labels = stroke_data_table.get_column(-1)
t0 = datetime.datetime.now()
dummy.fit(attr_data, class_labels)
t1 = datetime.datetime.now() - t0
print("Time elapsed for fit method:", t1)
t0 = datetime.datetime.now()
dummy_preds = dummy.predict(attr_data)
t1 = datetime.datetime.now() - t0
print("Time elapsed for predict method:", t1)
print("Number of predictions made (equal to the number of instances in the dataset:", len(dummy_preds))
print(dummy_preds)
```

All of the predictions are 0, since this dummy model simply chooses the most frequent of the 2 class labels and predicts that for every instance in the dataset. I then used those predictions to compute accuracy, precision, recall and f1 for this model.

```python
accuracy = evaluations.compute_accuracy(class_labels, dummy_preds)
print("accuracy:", accuracy)
precision = evaluations.compute_precision(class_labels, dummy_preds)
print("precision:", precision)
recall = evaluations.compute_recall(class_labels, dummy_preds)
print("recall:", recall)
f1 = evaluations.compute_f1(class_labels, dummy_preds)
print("f1", f1)
labels = ['0', '1']
confusion_matrix = evaluations.confusion_matrix(class_labels, dummy_preds, labels.copy())
evaluations.print_confusion_matrix("Dummy Model", labels.copy(), confusion_matrix)
```
